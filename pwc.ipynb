```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x = df.drop('KWH_Consumed' ,axis=1)
y= df['KWH_Consumed']
```

```python
x_rest , x_val , y_rest , y_val= train_test_split(x , y , test_size=0.1 , random_state=11)
```

```python
# validation_dataset
```

```python
validation_dataset = pd.concat([x_val,y_val] , axis=1)
validation_dataset
validation_dataset.to_csv('validation_dataset.csv', index=False)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_rest , y_rest , test_size=0.2 , random_state=11)
```

```python
from sklearn.metrics import mean_squared_error as MSE , mean_absolute_error as MAE, r2_score

def model_fit(string ,model , x_train , y_train , x_test , y_test , x_val , y_val):
    print(string.center(50,'*'))
    model.fit(x_train , y_train)
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    y_val_pred = model.predict(x_val)

    return y_train_pred , y_test_pred , y_val_pred

def eval (string , act_target , pred_target):
    print(string.center(50,'*'))
    mse = MSE(act_target ,pred_target)
    mae = MAE(act_target ,pred_target)
    R2Score= r2_score(act_target ,pred_target)
    print('MSE is :' , mse)
    print('MAE is :' , mae)
    print('r2score is :' , R2Score)
    
    


    
    
    
```

```python
lg_model = LinearRegression()
y_train_pred , y_test_pred , y_val_pred = model_fit('Linear Regression model',lg_model , x_train , y_train , x_test , y_test , x_val , y_val)
eval('evaluation on train_data' , y_train , y_train_pred)
eval('evaluation on test_data' , y_test , y_test_pred)
eval('evaluation on validation_data' , y_val , y_val_pred)
```

```python
rg_model = RandomForestRegressor()
y_train_pred , y_test_pred , y_val_pred = model_fit('Randomforest Regression model',rg_model , x_train , y_train , x_test , y_test , x_val , y_val)
eval('evaluation on train_data' , y_train , y_train_pred)
eval('evaluation on test_data' , y_test , y_test_pred)
eval('evaluation on test_data' , y_val , y_val_pred)

```
